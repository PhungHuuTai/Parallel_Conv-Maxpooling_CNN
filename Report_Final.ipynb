## <span style='color:#004080'>2.1. Ph√¢n t√≠ch </span>
·ªû ƒë·ªì √°n n√†y nh√≥m s·∫Ω ti·∫øn h√†nh build l·∫°i Class Conv2D v√† MaxPooling2D c·ªßa th∆∞ vi·ªán keras.

ƒêa s·ªë c√°c h√†m nh√≥m s·∫Ω s·ª≠ d·ª•ng l·∫°i t·ª´ trong source code t·ª´ th∆∞ vi·ªán. Nh√≥m ch·ªâ ti·∫øn h√†nh vi·∫øt l·∫°i h√†m convolution trong class Conv2D v√† maxpooling trong class MaxPooling2D ƒë·ªÉ ph√π h·ª£p v·ªõi m·ª•c ƒë√≠ch c·ªßa nh√≥m trong ƒë·ªì √°n n√†y.

### H√†m chu·∫©n h√≥a c√°c gi√° tr·ªã ƒë·∫ßu v√†o v·ªÅ chu·∫©n d·∫°ng tuple. Vd: 1 s·∫Ω cho ra (1,1)

```python
def standardize_tuple(value, n, name, allow_zero=False):
    error_msg = (
        f"The `{name}` argument must be a tuple of {n} integers. "
        f"Received {name}={value}"
    )

    if isinstance(value, int):
        value_tuple = (value,) * n
    else:
        try:
            value_tuple = tuple(value)
        except TypeError:
            raise ValueError(error_msg)
        if len(value_tuple) != n:
            raise ValueError(error_msg)
        for single_value in value_tuple:
            try:
                int(single_value)
            except (ValueError, TypeError):
                error_msg += (
                    f"including element {single_value} of "
                    f"type {type(single_value)}"
                )
                raise ValueError(error_msg)

    if allow_zero:
        unqualified_values = {v for v in value_tuple if v < 0}
        req_msg = ">= 0"
    else:
        unqualified_values = {v for v in value_tuple if v <= 0}
        req_msg = "> 0"

    if unqualified_values:
        error_msg += (
            f", including values {unqualified_values}"
            f" that do not satisfy `value {req_msg}`"
        )
        raise ValueError(error_msg)

    return value_tuple
```

### H√†m khoanh v√πng gi√° tr·ªã cho padding 

```python
def standardize_padding(value, allow_causal=False):
    if isinstance(value, (list, tuple)):
        return value
    padding = value.lower()
    if allow_causal:
        allowed_values = {"valid", "same", "causal"}
    else:
        allowed_values = {"valid", "same"}
    if padding not in allowed_values:
        raise ValueError(
            "The `padding` argument must be a list/tuple or one of "
            f"{allowed_values}. "
            f"Received: {padding}"
        )
    return padding
```

### H√†m khoanh v√πng gi√° tr·ªã cho data_format

```python
def standardize_data_format(data_format):
    if data_format is None:
        return 'channels_last'
    data_format = str(data_format).lower()
    if data_format not in {"channels_first", "channels_last"}:
        raise ValueError(
            "The `data_format` argument must be one of "
            "{'channels_first', 'channels_last'}. "
            f"Received: data_format={data_format}"
        )
    return data_format
```

### H√†m t√≠nh k√≠ch th∆∞·ªõc ƒë·∫ßu ra sau khi ƒëi qua l·ªõp Convolution

```python
def compute_output_conv_shape( 
        input_shape,
        filters,
        kernel_size,
        strides=1,
        padding="valid",
        data_format="channels_last",
        dilation_rate=1,
    ):
        """Compute the output shape of conv ops."""
        if data_format == "channels_last":
            spatial_shape = input_shape[1:-1]
            kernel_shape = kernel_size + (input_shape[-1], filters)
        else:
            spatial_shape = input_shape[2:]
            kernel_shape = kernel_size + (input_shape[1], filters)
        if len(kernel_shape) != len(input_shape):
            raise ValueError(
                "Kernel shape must have the same length as input, but received "
                f"kernel of shape {kernel_shape} and "
                f"input of shape {input_shape}."
            )
        if isinstance(dilation_rate, int):
            dilation_rate = (dilation_rate,) * len(spatial_shape)
        if isinstance(strides, int):
            strides = (strides,) * len(spatial_shape)
        if len(dilation_rate) != len(spatial_shape):
            raise ValueError(
                "Dilation must be None, scalar or tuple/list of length of "
                "inputs' spatial shape, but received "
                f"`dilation_rate={dilation_rate}` and "
                f"input of shape {input_shape}."
            )
        none_dims = []
        spatial_shape = np.array(spatial_shape)
        for i in range(len(spatial_shape)):
            if spatial_shape[i] is None:
                # Set `None` shape to a manual value so that we can run numpy
                # computation on `spatial_shape`.
                spatial_shape[i] = -1
                none_dims.append(i)

        kernel_spatial_shape = np.array(kernel_shape[:-2])
        dilation_rate = np.array(dilation_rate)
        if padding == "valid":
            output_spatial_shape = (
                np.floor(
                    (spatial_shape - dilation_rate * (kernel_spatial_shape - 1) - 1)
                    / strides
                )
                + 1
            )
            for i in range(len(output_spatial_shape)):
                if i not in none_dims and output_spatial_shape[i] < 0:
                    raise ValueError(
                        "Computed output size would be negative. Received "
                        f"`inputs shape={input_shape}`, "
                        f"`kernel shape={kernel_shape}`, "
                        f"`strides = {strides}`, "
                        f"`dilation_rate={dilation_rate}`."
                    )
        elif padding == "same" or padding == "causal":
            output_spatial_shape = np.floor((spatial_shape - 1) / strides) + 1
        else:
            raise ValueError(
                "`padding` must be either `'valid'` or `'same'`. Received "
                f"{padding}."
            )
        output_spatial_shape = [int(i) for i in output_spatial_shape]
        for i in none_dims:
            output_spatial_shape[i] = None
        output_spatial_shape = tuple(output_spatial_shape)
        if data_format == "channels_last":
            output_shape = (
                (input_shape[0],) + output_spatial_shape + (kernel_shape[-1],)
            )
        else:
            output_shape = (input_shape[0], kernel_shape[-1]) + output_spatial_shape
        return output_shape
```

### H√†m t√≠nh k√≠ch th∆∞·ªõc ƒë·∫ßu ra sau khi ƒëi qua l·ªõp MaxPooling

```python
def compute_pooling_output_shape(
    input_shape,
    pool_size,
    strides,
    padding="valid",
    data_format="channels_last",
):
    strides = pool_size if strides is None else strides
    input_shape_origin = list(input_shape)
    input_shape = np.array(input_shape)
    if data_format == "channels_last":
        spatial_shape = input_shape[1:-1]
    else:
        spatial_shape = input_shape[2:]
    none_dims = []
    for i in range(len(spatial_shape)):
        if spatial_shape[i] is None:
            # Set `None` shape to a manual value so that we can run numpy
            # computation on `spatial_shape`.
            spatial_shape[i] = -1
            none_dims.append(i)
    pool_size = np.array(pool_size)
    if padding == "valid":
        output_spatial_shape = (
            np.floor((spatial_shape - pool_size) / strides) + 1
        )
        for i in range(len(output_spatial_shape)):
            if i not in none_dims and output_spatial_shape[i] < 0:
                raise ValueError(
                    "Computed output size would be negative. Received: "
                    f"`inputs.shape={input_shape}` and `pool_size={pool_size}`."
                )
    elif padding == "same":
        output_spatial_shape = np.floor((spatial_shape - 1) / strides) + 1
    else:
        raise ValueError(
            "Argument `padding` must be either 'valid' or 'same'. Received: "
            f"padding={padding}"
        )
    output_spatial_shape = [int(i) for i in output_spatial_shape]
    for i in none_dims:
        output_spatial_shape[i] = None
    output_spatial_shape = tuple(output_spatial_shape)
    if data_format == "channels_last":
        output_shape = (
            (input_shape_origin[0],)
            + output_spatial_shape
            + (input_shape_origin[-1],)
        )
    else:
        output_shape = (
            input_shape_origin[0],
            input_shape_origin[1],
        ) + output_spatial_shape
    return output_shape
```

### <span style='color:#cc7a00'>2.3.1 Convolution </span>

```python
def create_patches(inputs, height_in, width_in, channel_in, height_kernel, width_kernel, strides, padding):
    pad_h, pad_w = padding
    out_h = (height_in + 2 * pad_h - height_kernel) // strides + 1
    out_w = (width_in + 2 * pad_w - width_kernel) // strides + 1

    # Padding if necessary
    if padding != (0, 0):
        inputs = tf.pad(inputs, [[0, 0], [pad_h, pad_h], [pad_w, pad_w], [0, 0]])

    # Extract image patches
    patches = tf.image.extract_patches(inputs, 
                                       sizes=[1, height_kernel, width_kernel, 1], 
                                       strides=[1, strides, strides, 1], 
                                       rates=[1, 1, 1, 1], 
                                       padding='VALID')

    patches = tf.reshape(patches, [-1, out_h * out_w, height_kernel * width_kernel * channel_in])
    return patches

def convolution(inputs, kernels, strides, padding, dilation_rate, data_format):
    input_shape = inputs.shape
    kernel_shape = kernels.shape
    batch_size, height_in, width_in, _ = input_shape
    
    output_shape = compute_output_conv_shape(input_shape, kernel_shape[-1], kernel_shape[:2], 
                                             strides=strides, padding=padding, 
                                             data_format=data_format, dilation_rate=dilation_rate)
    _, height_out, width_out, channel_out = output_shape

    patches_inputs = create_patches(inputs, height_in, width_in, input_shape[3], 
                          kernel_shape[0], kernel_shape[1], strides[0], (0, 0))
    kernels_2D = tf.reshape(kernels, [-1, kernel_shape[-1]])
    outputs_np = np.zeros((batch_size, height_out, width_out, channel_out), dtype=np.float32)
    
    start = time.time()
    for b in range(batch_size):
        inputs_np = patches_inputs[b].numpy()
        kernels_np = kernels_2D.numpy()
        result = inputs_np @ kernels_np
        result = np.reshape(result, [height_out, width_out, -1])
        outputs_np[b] = result
    end = time.time()
    print(f"Runtime Conv: {end-start}")
#     conv_result = tf.nn.convolution(inputs, kernels, strides=[1, 1, 1, 1], padding='VALID')
#     print(conv_result-outputs_np)
    outputs_ts = tf.convert_to_tensor(outputs_np)
    return outputs_ts
```

### X√¢y d·ª±ng class BaseConv

```python
class BaseConv(Layer):
    def __init__(self, rank, filters, kernel_size, strides=1, padding='valid',
                 data_format=None, dilation_rate=1, groups=1,
                 activation=None, use_bias=True, version='cpu',
                 kernel_initializer='glorot_uniform',
                 bias_initializer='zeros', kernel_regularizer=None,
                 bias_regularizer=None, activity_regularizer=None,
                 kernel_constraint=None, bias_constraint=None, trainable=True, name=None, **kwargs):
        super().__init__(
            trainable=trainable,
            name=name,
            activity_regularizer=activity_regularizer,
            **kwargs,
        )
        self.rank = rank
        self.filters = filters
        self.groups = groups
        self.kernel_size = standardize_tuple(kernel_size, rank, 'kernel_size')
        self.strides = standardize_tuple(strides, rank, 'strides')
        self.padding = standardize_padding(padding, allow_causal=rank == 1)
        self.data_format = standardize_data_format(data_format)
        self.dilation_rate = standardize_tuple(dilation_rate, rank, 'dilation_rate')
        self.activation = activations.get(activation)
        self.use_bias = use_bias
        self.version = version
        self.kernel_initializer = initializers.get(kernel_initializer)
        self.bias_initializer = initializers.get(bias_initializer)
        self.kernel_regularizer = regularizers.get(kernel_regularizer)
        self.bias_regularizer = regularizers.get(bias_regularizer)
        self.kernel_constraint = constraints.get(kernel_constraint)
        self.bias_constraint = constraints.get(bias_constraint)
        self.input_spec = InputSpec(min_ndim=self.rank + 2)
        self.data_format = self.data_format

    def build(self, input_shape):
        # determine the shape of the input volume
        if self.data_format == 'channels_last':
            channel_axis = -1
            input_channel = input_shape[-1]
        else:
            channel_axis = 1
            input_channel = input_shape[1]
        self.input_spec = InputSpec(
            min_ndim=self.rank + 2, axes={channel_axis: input_channel}
        )

        # calculate the shape of the kernel volume
        kernel_shape = self.kernel_size + (input_channel//self.groups, self.filters)

        # create the weight tensor
        self.kernel = self.add_weight(
            name="kernel",
            shape=kernel_shape,
            initializer=self.kernel_initializer,
            regularizer=self.kernel_regularizer,
            constraint=self.kernel_constraint,
            trainable=True,
            dtype=self.dtype,
        )

        # create the bias tensor
        if self.use_bias:
            self.bias = self.add_weight(
                name="bias",
                shape=(self.filters,),
                initializer=self.bias_initializer,
                regularizer=self.bias_regularizer,
                constraint=self.bias_constraint,
                trainable=True,
                dtype=self.dtype,
            )
        else:
            self.bias = None

        # calculate the shape of the output volume
        self.compute_output_shape(input_shape)
        self.built = True
    
    def call(self, inputs):
        # convolve the input volume with the filter
        if self.version == 'cpu':
            print('Host')
            outputs = convolution(
                inputs, 
                self.kernel, 
                strides=list(self.strides), 
                padding=self.padding, 
                dilation_rate=self.dilation_rate, 
                data_format=self.data_format,
            )
        elif self.version == 'gpu_base':
            print('VB')
            outputs = convolution_gpu_base(
                inputs, 
                self.kernel, 
                strides=list(self.strides), 
                padding=self.padding, 
                dilation_rate=self.dilation_rate, 
                data_format=self.data_format,
            )
        elif self.version == 'gpu_v1':
            print('V1')
            outputs = convolution_gpu_v1(
                inputs, 
                self.kernel, 
                strides=list(self.strides), 
                padding=self.padding, 
                dilation_rate=self.dilation_rate, 
                data_format=self.data_format,
            )
        elif self.version == 'gpu_v2':
            print('V2')
            outputs = convolution_gpu_v2(
                inputs, 
                self.kernel, 
                strides=list(self.strides), 
                padding=self.padding, 
                dilation_rate=self.dilation_rate, 
                data_format=self.data_format,
            )
        elif self.version == 'gpu_v3':
            print('V3')
            outputs = convolution_gpu_v3(
                inputs, 
                self.kernel, 
                strides=list(self.strides), 
                padding=self.padding, 
                dilation_rate=self.dilation_rate, 
                data_format=self.data_format,
            )
        
        if self.use_bias:
            if self.data_format == "channels_last":
                bias_shape = (1,) * (self.rank + 1) + (self.filters,)
            else:
                bias_shape = (1, self.filters) + (1,) * self.rank
            bias = ops.reshape(self.bias, bias_shape)
            outputs += bias
            
        # apply the activation function
        if self.activation != None:
            outputs = self.activation(outputs)

        return outputs
    
    def compute_output_shape(self, input_shape):
        return compute_output_conv_shape(
            input_shape,
            self.filters,
            self.kernel_size,
            strides = self.strides,
            padding = self.padding,
            data_format = self.data_format,
            dilation_rate = self.dilation_rate,
        )
    
    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "filters": self.filters,
                "kernel_size": self.kernel_size,
                "strides": self.strides,
                "padding": self.padding,
                "data_format": self.data_format,
                "dilation_rate": self.dilation_rate,
                "groups": self.groups,
                "activation": activations.serialize(self.activation),
                "use_bias": self.use_bias,
                "kernel_initializer": initializers.serialize(
                    self.kernel_initializer
                ),
                "bias_initializer": initializers.serialize(
                    self.bias_initializer
                ),
                "kernel_regularizer": regularizers.serialize(
                    self.kernel_regularizer
                ),
                "bias_regularizer": regularizers.serialize(
                    self.bias_regularizer
                ),
                "activity_regularizer": regularizers.serialize(
                    self.activity_regularizer
                ),
                "kernel_constraint": constraints.serialize(
                    self.kernel_constraint
                ),
                "bias_constraint": constraints.serialize(self.bias_constraint),
            }
        )
        return config
```

```python
class Conv2D(BaseConv):
    def __init__(
        self,
        filters,
        kernel_size,
        strides=(1, 1),
        padding="valid",
        data_format=None,
        dilation_rate=(1, 1),
        groups=1,
        activation=None,
        use_bias=True,
        version='cpu',
        kernel_initializer="glorot_uniform",
        bias_initializer="zeros",
        kernel_regularizer=None,
        bias_regularizer=None,
        activity_regularizer=None,
        kernel_constraint=None,
        bias_constraint=None,
        **kwargs
    ):
        super().__init__(
            rank=2,
            filters=filters,
            kernel_size=kernel_size,
            strides=strides,
            padding=padding,
            data_format=data_format,
            dilation_rate=dilation_rate,
            groups=groups,
            activation=activation,
            use_bias=use_bias,
            version=version,
            kernel_initializer=kernel_initializer,
            bias_initializer=bias_initializer,
            kernel_regularizer=kernel_regularizer,
            bias_regularizer=bias_regularizer,
            activity_regularizer=activity_regularizer,
            kernel_constraint=kernel_constraint,
            bias_constraint=bias_constraint,
            **kwargs
        )
```

### <span style='color:#cc7a00'>2.3.2 MaxPooling</span>

```python
def maxpooling(inputs, pool_size, strides, padding, data_format):
    input_shape = inputs.shape
    batch_size, height_in, width_in, channel_in = input_shape
    ps = pool_size[0]
    st = strides[0]
    
    _, height_out, width_out, channel_out = compute_pooling_output_shape(input_shape, pool_size, 
                                                                         strides, padding, data_format)
    inputs_np = inputs.numpy()
    outputs_np = np.zeros((batch_size, height_out, width_out, channel_out), dtype=np.float32)

    # T√≠nh max pooling
    start = time.time()
#     for b in range(batch_size):
    for i in range(height_out):
        for j in range(width_out):
            # L·∫•y gi√° tr·ªã max t·ª´ v√πng pooling
            pooling_region = inputs[:, i*st:i*st+ps, j*st:j*st+ps, :]
            outputs_np[:, i, j, :] = np.max(pooling_region, axis=(1,2))
    end = time.time()
    outputs_ts = tf.convert_to_tensor(outputs_np)
    print(f"Runtime MaxPooling: {end-start}")
#     print(tf.nn.max_pool(inputs, ksize=ps, strides=st, padding="VALID")-outputs_ts)
    return outputs_ts
```

```python
class BasePooling(Layer):
    def __init__(
        self,
        pool_size,
        strides,
        pool_dimensions,
        pool_mode="max",
        padding="valid",
        data_format=None,
        version='cpu',
        name=None,
        **kwargs,
    ):
        super().__init__(name=name, **kwargs)

        self.pool_size = standardize_tuple(
            pool_size, pool_dimensions, "pool_size"
        )
        strides = pool_size if strides is None else strides
        self.strides = standardize_tuple(
            strides, pool_dimensions, "strides", allow_zero=True
        )
        self.pool_mode = pool_mode
        self.padding = padding
        self.data_format = standardize_data_format(data_format)
        self.version = version
        self.input_spec = InputSpec(ndim=pool_dimensions + 2)

    def call(self, inputs):
        if self.version == 'cpu':
            print("Host")
            return maxpooling(
                inputs,
                pool_size=self.pool_size,
                strides=self.strides,
                padding=self.padding,
                data_format=self.data_format,
            )
        elif self.version == 'gpu_base':
            print("Base")
            return maxpooling_gpu_base(
                inputs,
                pool_size=self.pool_size,
                strides=self.strides,
                padding=self.padding,
                data_format=self.data_format,
            )
        elif self.version == 'gpu_v1':
            print("Version 1")
            return maxpooling_gpu_v1(
                inputs,
                pool_size=self.pool_size,
                strides=self.strides,
                padding=self.padding,
                data_format=self.data_format,
            )
        elif self.version == 'gpu_v2':
            print("Version 2")
            return maxpooling_gpu_v2(
                inputs,
                pool_size=self.pool_size,
                strides=self.strides,
                padding=self.padding,
                data_format=self.data_format,
            )
        elif self.version == 'gpu_v3':
            print("Version 3")
            return maxpooling_gpu_v3(
                inputs,
                pool_size=self.pool_size,
                strides=self.strides,
                padding=self.padding,
                data_format=self.data_format,
            )

    def compute_output_shape(self, input_shape):
        return compute_pooling_output_shape(
            input_shape,
            self.pool_size,
            self.strides,
            self.padding,
            self.data_format,
        )

    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "pool_size": self.pool_size,
                "padding": self.padding,
                "strides": self.strides,
                "data_format": self.data_format,
            }
        )
        return config
```

```python
class MaxPooling2D(BasePooling):
    def __init__(
        self,
        pool_size=(2, 2),
        strides=None,
        padding="valid",
        data_format=None,
        version='cpu',
        name=None,
        **kwargs
    ):
        super().__init__(
            pool_size,
            strides,
            pool_dimensions=2,
            pool_mode="max",
            padding=padding,
            data_format=data_format,
            version=version,
            name=name,
            **kwargs,
        )
```

## Architecture of image classification model

1. Convolutional Layer with 32 filters and a (3,3) kernel, followed by BatchNormalization, MaxPooling and Dropout
2. Convolutional Layer with 64 filters and a (3,3) kernel, followed by BatchNormalization, MaxPooling and Dropout
3. Convolutional Layer with 128 filters and a (3,3) kernel, followed by BatchNormalization, MaxPooling and Dropout
4. Flatten layer to flatten the output from the previous layers.
5. Dense layer with 512 units and ReLU activation, followed by BatchNormalization and Dropout.
6. Output layer with 1 unit and a sigmoid activation function for binary classification.

I'm using BatchNormalization and Dropout after convolutional and dense layers helps with regularization and generalization. The model is compiled with binary cross-entropy loss and the Adam optimizer.

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, 1)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

#### Convolution

```python
@cuda.jit
def mul_mat_kernel(A, B, C):
    c, r = cuda.grid(2)
    
    if r < C.shape[0] and c < C.shape[1]:
        temp = 0
        for i in range(A.shape[1]):
            temp += A[r, i] * B[i, c]
        C[r, c] = temp
        
# G·ªçi mul_mat_kernel v·ªõi m·∫£ng int nh·ªè ƒë·ªÉ c√≥ mul_mat_kernel ƒë√£ ƒë∆∞·ª£c bi√™n d·ªãch
# int_arr = np.empty((1, 1), dtype=int)
# mul_mat_kernel[1, 1](int_arr, int_arr, int_arr)
```

```python
def convolution_gpu_base(inputs, kernels, strides, padding, dilation_rate, data_format):
    input_shape = inputs.shape
    kernel_shape = kernels.shape
    batch_size, height_in, width_in, _ = input_shape
    
    output_shape = compute_output_conv_shape(input_shape, kernel_shape[-1], kernel_shape[:2], 
                                             strides=strides, padding=padding, 
                                             data_format=data_format, dilation_rate=dilation_rate)
    _, height_out, width_out, channel_out = output_shape
    
    patches_inputs = create_patches(inputs, height_in, width_in, input_shape[3], 
                          kernel_shape[0], kernel_shape[1], strides[0], (0, 0))
    kernels_2D = tf.reshape(kernels, [-1, kernel_shape[-1]])
    
    inputs_np = patches_inputs.numpy()
    kernels_np = kernels_2D.numpy()
    output_tensor = tf.TensorArray(dtype=tf.float32, size=0, dynamic_size=True)
    
    start = time.time()
    for b in range(batch_size):
        result = np.zeros((height_out*width_out, channel_out), dtype=np.float32)
        block_size = (32, 32)
        grid_size = (math.ceil(result.shape[1] / block_size[0]),
                     math.ceil(result.shape[0] / block_size[1]))
        mul_mat_kernel[grid_size, block_size](inputs_np[b], kernels_np, result)
        
        result = tf.reshape(result, [height_out, width_out, -1])
        output_tensor = output_tensor.write(b, result)
    end = time.time()
    print(f"Run time Conv: {end-start}")
#     conv_result = tf.nn.convolution(inputs, kernels, strides=[1, 1, 1, 1], padding='VALID')
#     print(conv_result - output_tensor.stack())
    return output_tensor.stack()
```

#### MaxPooling

```python
@cuda.jit
def maxpooling_kernel(A, B, strides):
    j, i = cuda.grid(2)
    
    in_i = i*strides
    in_j = j*strides
    
    if i<B.shape[0] and j<B.shape[1]:
        max_val = A[in_i, in_j]
        max_val = max(max_val, A[in_i, in_j + 1])
        max_val = max(max_val, A[in_i + 1, in_j])
        max_val = max(max_val, A[in_i + 1, in_j + 1])
        B[i, j] = max_val
```

```python
def maxpooling_gpu_base(inputs, pool_size, strides, padding, data_format):
    input_shape = inputs.shape
    batch_size, height_in, width_in, channel_in = input_shape
    ps = pool_size[0]
    st = strides[0]
    
    _, height_out, width_out, channel_out = compute_pooling_output_shape(input_shape, pool_size, 
                                                                         strides, padding, data_format)
    inputs_np = inputs.numpy()
    outputs_np = np.zeros((batch_size, height_out, width_out, channel_out), dtype=np.float32)
    
    start = time.time()
    for b in range(batch_size):
        for c in range(channel_out):
            result = np.zeros((height_out, width_out), dtype=np.float32)
            block_size = (32, 32)
            grid_size = (math.ceil(result.shape[1] / block_size[0]),
                         math.ceil(result.shape[0] / block_size[1]))
            maxpooling_kernel[grid_size, block_size](np.ascontiguousarray(inputs_np[b, :, :, c]), result, st)
            outputs_np[b, :, :, c] = result
    end = time.time()
    print(f"Runtime MaxPooling: {end-start}")
    
    outputs_ts = tf.convert_to_tensor(outputs_np)
#     print(tf.nn.max_pool(inputs, ksize=ps, strides=st, padding="VALID")-outputs_ts)
    return outputs_ts
```

```python
@cuda.jit
def mul_mat_kernel_v2(A, B, C): # Code c·ªßa 1 thread b√™n GPU/device
    c_B = cuda.const.array_like(B)
    c, r = cuda.grid(2)

    if r < C.shape[0] and c < C.shape[1]:
        temp = 0
        for i in range(A.shape[1]):
            temp += A[r, i] * c_B[i, c]
        C[r, c] = temp
        
# # G·ªçi mul_mat_kernel v·ªõi m·∫£ng int nh·ªè ƒë·ªÉ c√≥ mul_mat_kernel ƒë√£ ƒë∆∞·ª£c bi√™n d·ªãch
# int_arr = np.empty((1, 1), dtype=int)
# mul_mat_kernel_v2[1, 1](int_arr, int_arr, int_arr)
```

```python
def convolution_gpu_v2(inputs, kernels, strides, padding, dilation_rate, data_format):
    input_shape = inputs.shape
    kernel_shape = kernels.shape
    batch_size, height_in, width_in, _ = input_shape
    
    output_shape = compute_output_conv_shape(input_shape, kernel_shape[-1], kernel_shape[:2], 
                                             strides=strides, padding=padding, 
                                             data_format=data_format, dilation_rate=dilation_rate)
    _, height_out, width_out, channel_out = output_shape
  
    patches_inputs = create_patches(inputs, height_in, width_in, input_shape[3], 
                          kernel_shape[0], kernel_shape[1], strides[0], (0, 0))
    kernels_2D = tf.reshape(kernels, [-1, kernel_shape[-1]])
    
    inputs_np = patches_inputs.numpy()
    kernels_np = kernels_2D.numpy()
    output_tensor = tf.TensorArray(dtype=tf.float32, size=0, dynamic_size=True)
    
    d_B = cuda.to_device(kernels_np)
    streams = [cuda.stream() for _ in range(batch_size)]
    start = time.time()
    for b in range(batch_size):
        d_A = cuda.to_device(inputs_np[b], stream=streams[b])
        d_C = cuda.device_array((height_out*width_out, channel_out), dtype=np.float32, stream=streams[b])
        block_size = (32, 32)
        grid_size = (math.ceil(d_C.shape[1] / block_size[0]),
                     math.ceil(d_C.shape[0] / block_size[1]))
        mul_mat_kernel_v2[grid_size, block_size, streams[b]](d_A, d_B, d_C)
        result = d_C.copy_to_host(stream=streams[b])
        result = tf.reshape(result, [height_out, width_out, -1])
        output_tensor = output_tensor.write(b, result)
    end = time.time()
    
    for stream in streams:
        stream.synchronize()
#     conv_result = tf.nn.convolution(inputs, kernels, strides=[1, 1, 1, 1], padding='VALID')
#     print(tf.round(conv_result - output_tensor.stack()))
    print(f"Run time Conv V2: {end-start}")
    return output_tensor.stack()
```

#### MaxPooling

```python
@cuda.jit
def maxpooling_kernel_v2(A, B, strides):
    j, i = cuda.grid(2)
    b = cuda.blockIdx.z
    
    in_i = i*strides
    in_j = j*strides
    
    if i<B.shape[1] and j<B.shape[2] and b<B.shape[0]:
        max_val = A[b, in_i, in_j]
        max_val = max(max_val, A[b, in_i, in_j + 1])
        max_val = max(max_val, A[b, in_i + 1, in_j])
        max_val = max(max_val, A[b, in_i + 1, in_j + 1])
        B[b, i, j] = max_val
```

```python
def maxpooling_gpu_v2(inputs, pool_size, strides, padding, data_format):
    input_shape = inputs.shape
    batch_size, height_in, width_in, channel_in = input_shape
    ps = pool_size[0]
    st = strides[0]
    
    _, height_out, width_out, channel_out = compute_pooling_output_shape(input_shape, pool_size, 
                                                                         strides, padding, data_format)
    inputs_np = inputs.numpy()
    outputs_np = np.zeros((batch_size, height_out, width_out, channel_out), dtype=np.float32)
    
    streams = [cuda.stream() for _ in range(channel_out)]
    start = time.time()
    for c in range(channel_out):
        d_A = cuda.to_device(np.ascontiguousarray(inputs_np[:, :, :, c]), stream=streams[c])
        d_C = cuda.device_array((batch_size, height_out, width_out), dtype=np.float32, stream=streams[c])
        block_size = (32, 32)
        grid_size = (math.ceil(d_C.shape[2] / block_size[0]),
                     math.ceil(d_C.shape[1] / block_size[1]),
                     batch_size)
        maxpooling_kernel_v2[grid_size, block_size, streams[c]](d_A, d_C, st)
        result = d_C.copy_to_host(stream=streams[c])
        outputs_np[:, :, :, c] = result
        
    for stream in streams:
        stream.synchronize()
    end = time.time()
    print(f"Runtime MaxPooling: {end-start}")
    
    outputs_ts = tf.convert_to_tensor(outputs_np)
#     print((tf.nn.max_pool(inputs, ksize=ps, strides=st, padding="VALID")-outputs_ts))
    return outputs_ts
```

```python
TILE_WIDTH = 32
```

```python
@cuda.jit
def mul_mat_kernel_v3(A, B, C):
    # Khai b√°o 2 m·∫£ng s_A v√† s_B trong b·ªô nh·ªõ SMEM c·ªßa block m√† thread thu·ªôc v·ªÅ
    s_A = cuda.shared.array((TILE_WIDTH, TILE_WIDTH), dtype=A.dtype)
    s_B = cuda.shared.array((TILE_WIDTH, TILE_WIDTH), dtype=A.dtype)
    c_B = cuda.const.array_like(B)
    
    tx, ty = cuda.threadIdx.x, cuda.threadIdx.y
    c, r = cuda.grid(2)
    batch_idx = cuda.blockIdx.z
    
    if batch_idx >= A.shape[0]:
        return

    temp = 0
    for phase in range(math.ceil(A.shape[2] / TILE_WIDTH)):
        if r < A.shape[1] and phase * TILE_WIDTH + tx < A.shape[2]:
            s_A[ty, tx] = A[batch_idx, r, phase * TILE_WIDTH + tx]
        else:
            s_A[ty, tx] = 0
        if phase * TILE_WIDTH + ty < c_B.shape[0] and c < c_B.shape[1]:
            s_B[ty, tx] = c_B[phase * TILE_WIDTH + ty, c]
        else:
            s_B[ty, tx] = 0

        cuda.syncthreads()

        for i in range(TILE_WIDTH):
            temp += s_A[ty, i] * s_B[i, tx]

        cuda.syncthreads()

    if r < C.shape[1] and c < C.shape[2]:
        C[batch_idx, r, c] = temp
        
# int_arr = np.empty((1, 1), dtype=int)
# int_arr_1 = np.empty((1, 1, 1), dtype=int)
# mul_mat_kernel_v3[1, 1](int_arr_1, int_arr, int_arr_1)
```

```python
def convolution_gpu_v3(inputs, kernels, strides, padding, dilation_rate, data_format):
    input_shape = inputs.shape
    kernel_shape = kernels.shape
    batch_size, height_in, width_in, _ = input_shape
    
    output_shape = compute_output_conv_shape(input_shape, kernel_shape[-1], kernel_shape[:2], 
                                             strides=strides, padding=padding, 
                                             data_format=data_format, dilation_rate=dilation_rate)
    _, height_out, width_out, channel_out = output_shape
    
    patches_inputs = create_patches(inputs, height_in, width_in, input_shape[3], 
                          kernel_shape[0], kernel_shape[1], strides[0], (0, 0))
    kernels_2D = tf.reshape(kernels, [-1, kernel_shape[-1]])
    
    inputs_np = patches_inputs.numpy()
    kernels_np = kernels_2D.numpy()
    output_tensor = tf.TensorArray(dtype=tf.float32, size=0, dynamic_size=True)
    
    d_B = cuda.to_device(kernels_np)
    d_A = cuda.to_device(inputs_np)
    d_C = cuda.device_array((batch_size, height_out * width_out, channel_out), dtype=np.float32)

    block_size = (32, 32)
    grid_size = (math.ceil(d_C.shape[2] / block_size[0]),
                 math.ceil(d_C.shape[1] / block_size[1]),
                 batch_size)

    start = time.time()
    mul_mat_kernel_v3[grid_size, block_size](d_A, d_B, d_C)
    result = d_C.copy_to_host()
    end = time.time()
    
    for b in range(batch_size):
        batch_result = tf.reshape(result[b], [height_out, width_out, -1])
        output_tensor = output_tensor.write(b, batch_result)
#     conv_result = tf.nn.convolution(inputs, kernels, strides=[1, 1, 1, 1], padding='VALID')
#     print(tf.round(conv_result - output_tensor.stack()))
    print(f"Run time Conv V3 with Streams: {end-start}")
    return output_tensor.stack()
```
